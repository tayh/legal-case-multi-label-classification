```python
import pandas as pd
```

```python
df = pd.read_excel('Dados.xlsx')
```

```python
ementas_list = df['Ementas'].to_list()
```

```python
ementa = []
for i in ementas_list:
    #print(i.split('.')[2])
    ementa.append(i.split('.'))
```

```python
print(len(ementa))
```

```python
ementa[2][0]
```

```python
first_subject = []
second_subject = []
four_subject = []
three_subject = []
for k in range(len(ementa)):
    try:
        first_subject.append(ementa[k][0])
        second_subject.append(ementa[k][1])
        three_subject.append(ementa[k][2])
        four_subject.append(ementa[k][3])
    except IndexError:
        first_subject.append(None)
        second_subject.append(None)
        three_subject.append(None)
        four_subject.append(None)
```

```python
four_subject
```

```python
df1 = pd.DataFrame(list(zip(first_subject, second_subject, three_subject, four_subject)), 
               columns =['0', '1', '2', '3']) 
```

```python
df1['0'].value_counts()
```

```python
writer_object = pd.ExcelWriter("Count.xlsx", engine ='xlsxwriter') 

df1.to_excel(writer_object, sheet_name ='Sheet1', index=False) 

workbook = writer_object.book 
worksheet = writer_object.sheets['Sheet1'] 

format1 = workbook.add_format({'num_format': '#0'}) 

worksheet.set_column('A:A', 18, format1)
worksheet.set_column('B:B', 18, format1)
worksheet.set_column('C:C', 50, format1)
worksheet.set_column('D:D', 90, format1)
worksheet.set_column('E:E', 30, format1)
worksheet.set_column('F:F', 35, format1)
worksheet.set_column('G:G', 18, format1)
worksheet.set_column('H:H', 35, format1)

writer_object.save() 
```
